```python
import pandas as pd
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.vectorstores import FAISS
from langchain.docstore.document import Document

# Load your CSV (replace with your file path if needed)
df = pd.read_csv('webtext.csv')

# Combine text and domain for better context (optional)
df['content'] = df['text'].fillna('') + " (Source: " + df['domain'].fillna('') + ")"

# Ensure all content is string type
df['content'] = df['content'].astype(str)

# Create Document objects for LangChain
documents = [Document(page_content=text) for text in df['content'].tolist()]

# Initialize embeddings model
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Create FAISS vector store from documents
vectorstore = FAISS.from_documents(documents, embeddings)

vectorstore = FAISS.from_documents(documents, embeddings)
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Create text-generation pipeline using GPT-2
pipe = pipeline("text-generation", model="gpt2", max_length=100)

# Wrap in LangChain LLM interface
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever())
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA

# Create text-generation pipeline with padding token and max_new_tokens for output length
pipe = pipeline(
    "text-generation",
    model="gpt2",
    pad_token_id=50256,
    max_new_tokens=150  # generate up to 150 tokens for the answer
)

llm = HuggingFacePipeline(pipeline=pipe)

qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever())

query = "How do I open a new browser window in Firefox?"
answer = qa_chain.invoke({"query": query})
print("Answer:", answer['result'])
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA

# Create text-generation pipeline with max_new_tokens and pad_token_id
pipe = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=150,
    pad_token_id=50256
)

llm = HuggingFacePipeline(pipeline=pipe)

# Use retriever with limited results (k=3 for concise answers)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

# Create RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever)

# Define your question here
query = "what is chatgpt"

# Run the query
answer = qa_chain.invoke({"query": query})

# Print the answer
print("Answer:", answer['result'])
```
